```python
# @title 2: Environment Diagnostics & Fast Install
import os
import torch

print("🖥️ Running System Diagnostics...")
if torch.cuda.is_available():
    gpu_name = torch.cuda.get_device_name(0)
    print(f"  -> GPU Detected: {gpu_name}")
    
    # Enable CUDA optimizations
    torch.backends.cudnn.benchmark = True
    torch.backends.cuda.matmul.allow_tf32 = True
    print("  -> Matmul TF32 Enabled.")
    
    if torch.cuda.is_bf16_supported():
        print("  -> BF16 Precision: Supported (Optimizing for L4/A100/H100)")
        os.environ['URVC_PRECISION'] = 'bf16'
    else:
        print("  -> FP16 Precision: Supported (Optimizing for T4/V100)")
        os.environ['URVC_PRECISION'] = 'fp16'
else:
    print("  -> ⚠️ Running on CPU mode. Training and generation will be slow.")

print("\n📦 Installing dependencies via uv...")
!apt install -y python3-dev unzip
!curl -LsSf https://astral.sh/uv/install.sh | sh
os.environ['PATH'] = '/root/.local/bin:' + os.environ['PATH']
os.environ['URVC_CONSOLE_LOG_LEVEL'] = 'WARNING'
os.environ['PYTHONIOENCODING'] = 'utf-8'
os.environ['PYTHONUTF8'] = '1'
os.environ['UV_COLOR'] = 'never'

if os.path.exists('uv.lock'):
    os.remove('uv.lock')

print("\n🚀 Building Virtual Environment & Initializing RVC Core...")
!uv run --prerelease if-necessary-or-explicit ./src/ultimate_rvc/core/main.py
!uv add --prerelease if-necessary-or-explicit matplotlib-inline==0.1.7

print("\n✨ Installation completed successfully!")
```
